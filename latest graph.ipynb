```python
##combining 
import os
from functools import partial
snr = "-15"
filename ="AF_c{}".format(snr)
os.mkdir(filename)
root_directory = '/Users/asamarka/paper/data/fusion/{}'.format(filename)  
list = ('test/8', 'test/16', 'test/32','train/8', 'train/16', 'train/32')
  
concat_root_path = partial(os.path.join, root_directory)
make_directory = partial(os.makedirs, exist_ok=True)
  
for path_items in map(concat_root_path, list):
    make_directory(path_items)

ml = {"train":350, "test":150}
modulation = [8,16,32] 
for m in ml:
    for j in modulation:
        root_1  = './{}AF/{}/{}'.format(snr,m,j)
        root_2= './{}C/{}/{}'.format(snr, m,j)
        for i in range(1,ml[m]+1):
            im1 = Image.open(os.path.join(root_1,'QAM{}.png'.format(i)))
            im2 = Image.open(os.path.join(root_2,'QAM{}.png'.format(i)))
            # 712
            _im1 = im1.resize((711, int(616 * 711 / 747)))
            _im2 = im2
            dst = Image.new('RGB', (_im1.width, _im1.height + 712))
            dst.paste(_im1, (0, 0))
            dst.paste(_im2, (0, _im1.height))
            dst.save('./AF_c{}/{}/{}/{}QAM.png'.format(snr,m,j,i))
```
